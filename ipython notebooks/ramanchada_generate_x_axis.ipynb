# *ramanchada*: Generate & assign new x axis to spectrum

```python
from ramanchada.classes import make_test_RamanChada
```

Load a Raman spetrum. It can have arbitrary values as an x axis (like channel numbers).

```python
r = make_test_RamanChada()
```

```python
r.reset_x()
```

```python
r.plot()
```

First, we'll have to find peaks in the spectrum

```python
r.peaks()
r.show_bands()
r.bands
```

Each found peak has an index. We can now generate a new x axis by using the method *make_x_axis* and telling it the true position in 1/cm for at least **two** of the peaks.

```python
xa = r.make_x_axis({0:1700, 8:150, 2:3500})
```

*xa* is the x_axis, stored as a *RamanCalibration* class instance.

```python
xa.show()
```

We can now store this x_axis (*.save()* method) and apply it to new spectra:

```python
r.assign_x(xa)
```

```python
r.log
```

```python
r.plot()
```
