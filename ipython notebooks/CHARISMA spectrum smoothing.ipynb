# CHARISMA - ramanchada spectrum smoothing

```python
from ramanchada.classes import make_test_RamanChada
```

Standard (no arguments): Savitzky-Golay filter

```python
s = make_test_RamanChada()
s.smooth()
s.plot()
```

Savitzky-Golay filter

```python
s = make_test_RamanChada()
s.smooth('sg', window_length=51, polyorder=3)
s.plot()
```

Wiener filter

```python
s = make_test_RamanChada()
s.smooth('wiener', mysize=51)
s.plot()
```

Median filter

```python
s = make_test_RamanChada()
s.smooth('median', kernel_size=51)
s.plot()
```

Gauss filter

```python
s = make_test_RamanChada()
s.smooth('gauss', sigma=51)
s.plot()
```

LOWESS filter

```python
s = make_test_RamanChada()
s.smooth('lowess', span=9)
s.plot()
```

Boxcar (no) filter

```python
s = make_test_RamanChada()
s.smooth('boxcar', box_pts=11)
s.plot()
```
