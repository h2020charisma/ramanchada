# CHARISMA: Spectrum peak fitting with the ramanchada library

```python
from ramanchada.classes import make_test_RamanChada
from ramanchada.analysis.peaks import voigt_fit, find_spectrum_peaks, double_voigt_fit, gauss_lorentz_fit, pearson4_fit, beta_profile_fit, lorentz_fit
```

Load test spectrum

```python
r_full = make_test_RamanChada()
```

```python
r_full.plot()
```

```python
r_full.peaks(fit=False)
r_full.bands
```

Voigt distribution

```python
r_full.peaks(fitmethod='voigt', interval_width=1.5, show=True, prominence=.2)
r_full.bands
```

Gaussian-Lorentzian sum

```python
r_full.peaks(fitmethod='gl', interval_width=1.5, show=True, prominence=.2)
r_full.bands
```

Pearson IV

```python
r_full.peaks(fitmethod='pearson', interval_width=1.5, show=True, prominence=.2)
r_full.bands
```

Lorentz

```python
r_full.peaks(fitmethod='lorentz', interval_width=1.5, show=True, prominence=.2)
r_full.bands
```

Beta distribution (won't fit well to this data)

```python
r_full.peaks(fitmethod='beta', interval_width=1.5, show=True, prominence=.2)
r_full.bands
```
